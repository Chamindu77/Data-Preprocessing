```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
# Load data
df = pd.read_csv('states.csv')

# Display first few raws
df.head()
```

```python
fig = px.scatter_mapbox(df, lat='latitude', lon = 'longitude', zoom = 4, mapbox_style = "open-street-map", title = 'Basic Distribution Center map' )
fig.show()
```

```python
np.random.seed(42)
df['sell_quantity'] = np.random.randint(100, 1000, size=len(df))
```

```python
fig = px.scatter_mapbox(
    df,
    lat="latitude",
    lon="longitude",
    hover_name="name",
    hover_data=["sell_quantity"],
    color="sell_quantity",
    zoom=4,
    mapbox_style="open-street-map",
    title="Basic Distribution Center Map"
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter_mapbox(
    df,
    lat="latitude",
    lon="longitude",
    hover_name="name",
    hover_data=["country_name", "state_code"],
    color="country_name",
    zoom=4,
    height=600,
    mapbox_style="open-street-map",
    title="Basic Distribution Center Map"
)

fig.show()
```
