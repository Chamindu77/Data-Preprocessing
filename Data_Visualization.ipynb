### Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
census_data = pd.read_csv('nyc_census_tracts.csv')

# Display first few raws
census_data.head()
```

## 1. Bar Plot

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x='Borough', y='TotalPop', data= census_data, palette="viridis") # palette="Set2"

# for container in ax.containers:
#     ax.bar_label(container, fmt='%.0f', rotation=45, label_type='edge')

plt.title('Total Population by Borough')
plt.xlabel('Borough')
plt.ylabel('Total Population')

plt.xticks(rotation=45)

plt.show()
```

## 2. Histogram

```python
plt.figure(figsize=(12,6))

sns.histplot(census_data['IncomePerCap'], bins=20, kde=True)

plt.title('Income Per Capita Distribution')
plt.xlabel('Income Per Capita')
plt.ylabel('Frequency')

plt.show()
```

### 3. Scatter Plot

```python
plt.figure(figsize=(12,6))

sns.scatterplot(x='Income', y='TotalPop', data=census_data, hue='Borough', palette='viridis')

plt.title('Income vs Population')
plt.xlabel('Income')
plt.ylabel('Total Population')

plt.show()
```

### 4. Boxplot

```python
 plt.figure(figsize=(12,6))

 sns.boxplot(x='Borough', y='Income', data=census_data, palette='viridis')

 plt.title('Income Destribution by Borough')
 plt.xlabel('Borough')
 plt.ylabel('Income')

 plt.xticks(rotation=45)

 plt.show()
```

### 5. Pie Chart

```python
gender_counts = [census_data['Men'].sum(), census_data['Women'].sum()]

labels = ['Men', 'Women']

plt.figure(figsize=(8,8))

colors = sns.color_palette("viridis", 2)
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)

plt.title('Gender Distribution in NY Census Data')

plt.show()
```

### 6. Violin Plot

```python
plt.figure(figsize=(12,6))

sns.violinplot(x= 'Borough',y='Income', data = census_data, palette = 'coolwarm' )

plt.title("Income Distribution by Borough", fontsize = 14, fontweight = 'bold' )
plt.xlabel("Borough", fontsize = 12)
plt.ylabel("Income", fontsize= 12)

plt.xticks(rotation = 45)

plt.show()
```

### 7. Heatmap

```python
plt.figure(figsize=(10,6))

corr_mattrix = census_data[["Income","IncomePerCap","ChildPoverty","Employed", "Unemployment"]]. corr()

sns.heatmap(corr_mattrix, annot = True, cmap = 'coolwarm', linewidths = 0.5)

plt.title("Correlation Heatmap of Economic Factors", fontsize = 14, fontweight ='bold')
plt.show()
```

### 8. Pairplot

```python
census_data.columns
```

```python
sns.pairplot(census_data[['Income','IncomePerCap','Poverty','Employed' ]], diag_kind = 'kde', palette='coolwarm')
plt.show()
```

### 9. Stacked Bar Chart

```python
plt.figure(figsize=(12,6))

gender_distribution = census_data.groupby('Borough')[['Men','Women']].sum()
gender_distribution.plot(kind = 'bar',stacked = True, colormap = 'viridis', figsize = (12,6))

plt.title("Gender Distribution Across Boroughs", fontsize = 14, fontweight = 'bold')
plt.xlabel('Borough',fontsize = 12)
plt.ylabel('Total Population', fontsize = 12)

plt.xticks(rotation = 45)

plt.legend(title = "Gender")
```

### 10. Donut Chart

```python
race_counts = [census_data['White'].mean(), census_data['Black'].mean(), census_data['Hispanic'].mean(), census_data['Asian'].mean(), census_data['Native'].mean()]
labels = ['White', 'Black', 'Hispanic', 'Asian', 'Native']

plt.figure(figsize=(6,6))

colors = sns.color_palette("viridis")
plt.pie(race_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'width': 0.5}, pctdistance=0.85)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Race Distribution in NY Census Data')

plt.show()
```
