### Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
census_data = pd.read_csv('nyc_census_tracts.csv')

# Display first few raws
census_data.head()
```

## 1. Bar Plot

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x='Borough', y='TotalPop', data= census_data, palette="viridis") # palette="Set2"

# for container in ax.containers:
#     ax.bar_label(container, fmt='%.0f', rotation=45, label_type='edge')

plt.title('Total Population by Borough')
plt.xlabel('Borough')
plt.ylabel('Total Population')

plt.xticks(rotation=45)

plt.show()
```

## 2. Histogram

```python
plt.figure(figsize=(12,6))

sns.histplot(census_data['IncomePerCap'], bins=20, kde=True)

plt.title('Income Per Capita Distribution')
plt.xlabel('Income Per Capita')
plt.ylabel('Frequency')

plt.show()
```

### 3. Scatter Plot

```python
plt.figure(figsize=(12,6))

sns.scatterplot(x='Income', y='TotalPop', data=census_data, hue='Borough', palette='viridis')

plt.title('Income vs Population')
plt.xlabel('Income')
plt.ylabel('Total Population')

plt.show()
```

### 4. Boxplot

```python
 plt.figure(figsize=(12,6))

 sns.boxplot(x='Borough', y='Income', data=census_data, palette='viridis')

 plt.title('Income Destribution by Borough')
 plt.xlabel('Borough')
 plt.ylabel('Income')

 plt.xticks(rotation=45)

 plt.show()
```

### 5. Pie Chart

```python
gender_counts = [census_data['Men'].sum(), census_data['Women'].sum()]

labels = ['Men', 'Women']

plt.figure(figsize=(8,8))

colors = sns.color_palette("viridis", 2)
plt.pie(gender_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)

plt.title('Gender Distribution in NY Census Data')

plt.show()
```
